```python
import numpy as np

epochs = 200
hidden_dim = 128
seq_len = 5
batch_size = 64
seed = 1

alpha_path = f"weights/DARNN_Figure3_L{seq_len}_HDE{hidden_dim}_HDD{hidden_dim}_BS{batch_size}_E{epochs}_s{seed}"

train_alpha = np.load(f"{alpha_path}_train_alpha.npy")
test_alpha = np.load(f"{alpha_path}_test_alpha.npy")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data for training and test sets
driving_series = np.arange(1, 163)  # Driving series indices (1 to 160)

# Create the figure and axes
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Plot for training set
axs[0].plot(driving_series, train_alpha, color='blue')
axs[0].set_title('NASDAQ 100 Training Set', fontsize=15)
axs[0].set_xlabel('Driving Series', fontsize=15)
axs[0].set_ylabel('Input Attention Weights', fontsize=15)
axs[0].grid(True, linestyle='--', alpha=0.6)

# Plot for test set
axs[1].plot(driving_series, test_alpha, color='blue')
axs[1].set_title('NASDAQ 100 Test Set', fontsize=15)
axs[1].set_xlabel('Driving Series', fontsize=15)
axs[1].grid(True, linestyle='--', alpha=0.6)

# Adjust layout and add scale
for ax in axs:
    ax.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))

plt.tight_layout()
plt.show()
```
